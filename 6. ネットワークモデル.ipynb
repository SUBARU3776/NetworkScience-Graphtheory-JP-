# ネットワークモデル
ネットワークの理解を深めるためには、まずネットワークがどのように構築され、その後どのように解析されるのかについての基本的なメカニズムを理解することが重要です。ネットワークは、ノード（要素）とエッジ（その間の相互作用）によって構成されます。例えば、ソーシャルメディアの友人関係やインターネットのウェブページ間のリンクなどがこれに該当します。

ネットワーク解析では、その構造や特性を明らかにするためにさまざまな手法が用いられます。ランダムグラフ、コンフィグレーションモデル、スケールフリーグラフ、スモールワールドグラフなどのモデルは、それぞれ異なる特性を持ち、現実世界のネットワークを理解するための手がかりを提供します。

これらのモデルは、ネットワークの特定の側面を強調し、理論的なフレームワークを提供します。例えば、**ランダムグラフは均一な接続性**を示し、**スケールフリーグラフはハブノードによる力学的な構造を持つ**ことが特徴です。これらのモデルを理解することで、ネットワークの形成や機能、情報の伝播パターンなどについて深く理解を深めることができます。

それぞれのグラフモデルには、次数分布、クラスタ係数、平均経路長、乗り換え確率において特徴的な違いがあります。以下に、それぞれの特徴を示します。<br>
**1. ランダムグラフ (Erdős-Rényiモデル)**<br>
**・** 次数分布：ポアソン分布に従う。すべてのノードが同じ確率でエッジを持ち、ノードの次数は平均的に均等。<br>
**・** クラスタ係数：低い。ランダムにエッジが追加されるため、ノードの近傍に三角形が形成されにくい。<br>
**・** 平均経路長：比較的短い。ノード間の距離は対数スケールで短くなる傾向がある。<br>
**・** 乗り換え確率：特に定義されていないが、一般的に均等にエッジが分布しているため、乗り換え確率は均等である。<br>

**2. コンフィグレーションモデル**<br>
**・** 次数分布：指定された次数分布に従う。任意の次数分布（例えば、パワー則など）を設定できる。<br>
**・** クラスタ係数：比較的低い。ノード間の接続は次数分布に依存するが、一般的にクラスタリングは低い。<br>
**・** 平均経路長：一般的に短い。ノード間の距離は対数スケールで短くなることが多い。<br>
**・** 乗り換え確率：特に定義されていないが、次数分布に従うため、ノード間の接続は次数に依存する。<br>

**3. BAモデル (Barabási-Albertモデル)**<br>
**・** 次数分布：スケールフリーネットワーク。次数分布はパワー則に従い、いくつかのノードが非常に高い次数を持ち、その他のノードは低い次数を持つ。<br>
**・** クラスタ係数：中程度。新しいノードが既存のノードに接続されるため、局所的にはある程度のクラスタリングが見られる。<br>
**・** 平均経路長：非常に短い。ネットワークはスケールフリーであり、ノード間の距離が小さくなる傾向がある。<br>
**・** 乗り換え確率：特に定義されていないが、次数の高いノードが優先的に接続されるため、特定のノードにエッジが集中する。<br>

**4. スモールワールド (Watts-Strogatzモデル)**<br>
**・** 次数分布：均等またはポアソン分布に近い。ノード数が大きくなると、次数分布は均等に近づくことが多い。<br>
**・** クラスタ係数：高い。ノード間に多くの三角形が形成され、局所的なクラスタリングが強い。<br>
**・** 平均経路長：短い。ノード間の距離が短く、スモールワールド特性を持つ。<br>
**・** 乗り換え確率：ランダム化の程度によって変わる。ノードの接続が部分的にランダムに置き換えられるため、乗り換え確率は調整可能。<br>

**5. 完全グラフ**<br>
**・** 次数分布：すべてのノードが同じ次数を持つ。ノード数が $N$ の場合、次数は $N−1$。<br>
**・** クラスタ係数：1。すべてのノードが完全に接続されているため、三角形は常に形成される。<br>
**・** 平均経路長：1。すべてのノードが直接つながっているため、距離は1。<br>
**・** 乗り換え確率：特に定義されていないが、すべてのノード間に直接エッジが存在するため、乗り換え確率は意味を持たない。<br>

## 次数分布の違いを確認してみましょう
ネットワークの次数分布は、ネットワーク内の各ノードが持つエッジ（つながり）の数に関する分布を指します。ノードの次数とは、そのノードに接続されているエッジの数を意味します。
具体的には、次数分布は以下のように理解できます：

1. **定義** <br>
・ ネットワーク内の各ノードについて、その次数（エッジの数）がどのように分布しているかを示す統計的な分布です。<br>

2. **重要性** <br>
・ 次数分布はネットワークの構造を理解する上で重要な役割を果たします。例えば、次数分布がどのような形状をしているかによって、ネットワークがランダムか、スケールフリーか、または他の特定のモデルにどれだけ近いかを推定することができます。<br>

3. **例** <br>
・ スケールフリーネットワークの場合、次数分布はべき乗則（power law）に従うことが一般的です。つまり、少数のハブノードが非常に高い次数を持ち、多数のノードが低い次数を持つ分布です。一方、ランダムグラフの場合、次数分布はポアソン分布に近くなります。<br>

4. **解析方法** <br>
・ ネットワークの次数分布を調べるためには、実際のネットワークデータから各ノードの次数を数え上げて分布をプロットする方法が一般的です。その後、得られた分布がどのような特性を持っているかを定量化することができます。<br>

次数分布の理解は、ネットワークがどのように形成され、情報が伝播するかを理解するための基礎となります。<br>

このスクリプトは、異なる次数分布を持つ4つの異なるネットワークを生成し、その構造と次数分布を比較するものです。比較しやすいように頂点数は同じですが、各ネットワークは異なる生成方法によって特徴的な構造を持っています。それぞれのネットワークの次数分布とその目的を説明します。

**スクリプトの目的**<br>
このスクリプトの目的は、異なる**ネットワーク生成モデルがどのような構造と次数分布を持つかを視覚的に比較することです。** それぞれのネットワークの次数分布をプロットすることで、各モデルの特徴やその影響を理解しやすくなります。これにより、実世界のネットワークをモデル化する際に適切な生成方法を選択する手助けとなります。<br>

**1. グラフの基本特性を理解する**<br>
・ 各グラフモデルがどのような構造を持ち、どのような特性を示すかを理解する。<br>

**2. 視覚化による比較**<br>
・ グラフの構造を視覚的に比較し、各モデルの特徴を明確にする。<br>

**3. 次数分布の分析**<br>
・ 各グラフの次数分布を分析し、特定のモデルがどのような分布を示すかを確認する。<br>


**各モデル**<br>
以下の4つのグラフモデルを扱います。<br>

**1. ランダムグラフ（Erdős-Rényi）**<br>
・ 生成方法: nx.erdos_renyi_graph(34, 0.1)<br>
・ 説明: ランダムグラフでは、各エッジが一定の確率で独立に存在します。このため、多くの頂点の次数は平均次数に近く、次数分布はポアソン分布に近づきます。<br>
・ 特徴: 各頂点が他の頂点とランダムに接続されるため、均一な次数分布を持ちます。<br>

**2. スケールフリーグラフ（Barabási-Albert）**<br>
・ 生成方法: nx.barabasi_albert_graph(34, 3)<br>
・ 説明: スケールフリーグラフでは、ネットワーク成長の過程で新しいノードが既存のノードに比例して接続されるため、少数のノードが非常に多くのエッジを持ち、多数のノードは比較的少数のエッジしか持ちません。<br>
・ 特徴: 次数分布はべき乗則に従い、長いテールを持ちます。<br>

**3. 完全グラフ（Complete）**<br>
・ 生成方法: nx.complete_graph(34)<br>
・ 説明: 完全グラフでは、すべてのノードが他のすべてのノードと接続されているため、各ノードの次数は全て等しく、次数分布は非常に狭いものになります。<br>
・ 特徴: 次数分布が一定で、全てのノードの次数が33（ノード数(34)-1）になります。<br>

**4. 空手クラブグラフ（Karate Club）**<br>
・ 生成方法: nx.karate_club_graph()<br>
・ 説明: 空手クラブグラフは、現実世界の社会ネットワークをモデル化したものであり、特定の平均次数を持ちます。クラブメンバー間の友情関係を表しています。<br>
・ 特徴: 実世界のネットワークに見られるようなクラスター構造やコミュニティ構造が含まれます。<br>
※ 空手クラブグラフは、実際の社会ネットワークの構造を反映しており、特定の平均次数を持ちます。実世界のネットワークの一例としてランダムグラフやスケールフリーグラフなどと比較する際に掲載しています。


**基本特性の出力**<br>
1. ノード数（Number of nodes）<br>
2. エッジ数（Number of edges）<br>
3. 平均次数（Average degree）<br>


**次数分布のプロット（スクリプトの流れ）**<br>
1. 各グラフモデルを生成する<br>
2. グラフを描画する<br>
3. 各グラフの基本的な特性をコンソールに出力する<br>
4. 各グラフの次数分布をプロットする<br>


```python
import networkx as nx
import matplotlib.pyplot as plt
# グラフの生成（ノード数はKarate Clubと比較しやすいように34に設定）
# ランダムグラフの生成（Erdős-Rényi グラフ）
er = nx.erdos_renyi_graph(34, 0.1)
plt.subplot(2, 4, 1)
nx.draw(er, node_size=10, node_color='red')
plt.title("Erdős-Rényi")
print("erdos_renyi_graph")
print("Number of nodes: ", nx.number_of_nodes(er))
print("Number of edges:", nx.number_of_edges(er))
print("Average degree: ", '{:.3f}'.format(nx.number_of_edges(er) * 2 / nx.number_of_nodes(er)))
plt.subplot(2, 4, 5)
plt.plot(nx.degree_histogram(er))
plt.title("Erdős-Rényi")
# 説明: ランダムグラフでは、各エッジが一定の確率で独立に存在するため、多くの頂点の次数は平均次数に近い。

# スケールフリーグラフの生成（Barabási-Albert グラフ）
ba = nx.barabasi_albert_graph(34, 3)
plt.subplot(2, 4, 2)
nx.draw(ba, node_size=10, node_color='red')
plt.title("Barabási-Albert")
print("barabasi_albert_graph")
print("Number of nodes: ", nx.number_of_nodes(ba))
print("Number of edges: ", nx.number_of_edges(ba))
print("Average degree: ", '{:.3f}'.format(nx.number_of_edges(ba) * 2 / nx.number_of_nodes(ba)))
plt.subplot(2, 4, 6)
plt.plot(nx.degree_histogram(ba))
plt.title("Barabási-Albert")
# 説明: スケールフリーグラフでは、少数のノードが非常に多くのエッジを持ち、多数のノードは比較的少数のエッジしか持たない。

# 完全グラフの生成
K_34 = nx.complete_graph(34)
plt.subplot(2, 4, 3)
nx.draw(K_34, node_size=10, node_color='red')
plt.title("Complete")
print("complete_graph")
print("Number of nodes: ", nx.number_of_nodes(K_34))
print("Number of edges: ", nx.number_of_edges(K_34))
print("Average degree: ", '{:.3f}'.format(nx.number_of_edges(K_34) * 2 / nx.number_of_nodes(K_34)))
plt.subplot(2, 4, 7)
plt.plot(nx.degree_histogram(K_34))
plt.title("Complete")
# 説明: 完全グラフでは、すべてのノードが他のすべてのノードと接続されているため、各ノードの次数は全て等しい。

# 空手クラブグラフの生成（実世界のネットワーク）
karate = nx.karate_club_graph()
plt.subplot(2, 4, 4)
nx.draw(karate, node_size=10, node_color='red')
plt.title("Karate Club")
print("karate_club_graph")
print("Number of nodes: ", nx.number_of_nodes(karate))
print("Number of edges: ", nx.number_of_edges(karate))
print("Average degree: ", '{:.3f}'.format(nx.number_of_edges(karate) * 2 / nx.number_of_nodes(karate)))
plt.subplot(2, 4, 8)
plt.plot(nx.degree_histogram(karate))
plt.title("Karate Club")
# 説明: 空手クラブグラフは、実際の社会ネットワークの構造を反映しており、特定の平均次数を持つ。
# 実世界のネットワークの一例として、ランダムグラフやスケールフリーグラフなどと比較する際に有用である。

plt.tight_layout()
plt.show()
```

## ランダムグラフ
**ランダムグラフはノード（頂点）とエッジ（辺）をランダムに生成して作成されるグラフです。**　この手法はネットワークの構造や特性を理解するための基礎的なモデルとして広く使われています。<br>

具体的には次のように作成されます。まず、一定数のノードを用意します。そして、各ペアのノード間にエッジを作成するかどうかを確率に基づいて決定します。例えば、コインを投げて表が出たらエッジを作成し、裏が出たらエッジを作成しない、といった具合です。このようにして、全てのノードペアについてエッジの有無を決定することで、ランダムグラフが生成されます。<br>

身近な例を挙げると、友達関係を考えてみましょう。あるクラスに10人の生徒がいるとします。この生徒たちが互いに友達になる確率が一定であると仮定すると、その友達関係をランダムグラフとしてモデル化できます。例えば、友達になる確率が0.3（30%）である場合、各生徒ペアについて、30%の確率で友達関係（エッジ）が形成されます。このようにして、クラス全体の友達関係がランダムに決まります。<br>

ランダムグラフの計算手順は次の通りです。まず、ノードの数を決定します。次に、各ノードペアについてエッジを作成する確率を決めます。この確率に基づいてランダムにエッジを生成し、全てのノードペアについてこの操作を繰り返します。例えば、エッジを作成する確率が0.3である場合、各ノードペアについて30%の確率でエッジを作成し、70%の確率でエッジを作成しません。これを全てのノードペアについて行うと、ランダムグラフが完成します。<br>

ランダムグラフの応用例としては、以下のようなものがあります。<br>

**1. ネットワークの基本特性の研究**：ランダムグラフは、ネットワークの基本的な性質（例えば、平均径、クラスタ係数、度分布など）を理解するための基本モデルとして使用されます。<br>
**2. 情報伝播のシミュレーション**：ソーシャルネットワーク上での情報やウイルスの伝播をシミュレーションする際に、ランダムグラフが使用されます。これにより、情報がどのように広がるかを予測することができます。<br>
**3. インターネットトラフィックのモデル化**：インターネット上のデータトラフィックをモデル化する際にもランダムグラフが使用されます。これにより、ネットワークの負荷やパフォーマンスを評価することができます。<br>
**4. 生物学的ネットワークの研究**：遺伝子やタンパク質の相互作用ネットワークの基本モデルとしてランダムグラフが使用されます。これにより、生物学的プロセスの基本的な構造を理解することができます。<br>

ランダムグラフは、ネットワークの基本的な特性を理解するための強力なツールであり、様々な分野で広く応用されています。<br>

```python
import networkx as nx
import matplotlib.pyplot as plt

# ERグラフの生成・Erdős-Rényiグラフ (ランダムグラフ)
er = nx.erdos_renyi_graph(100, 0.1)

# グラフの描画
plt.subplot(1, 2, 1)
nx.draw(er, node_size=10, node_color='red')
plt.title("Erdős-Rényi Graph")

# 基本特性の出力
print("erdos_renyi_graph")
print("Number of nodes: ", nx.number_of_nodes(er))
print("Number of edges:", nx.number_of_edges(er))
print("Average degree: ", '{:.3f}'.format(nx.number_of_edges(er) * 2 / nx.number_of_nodes(er)))

# 次数分布のプロット
plt.subplot(1, 2, 2)
plt.plot(nx.degree_histogram(er))
plt.title("Degree Distribution")
plt.show()
```

## ポアソン分布について
ポアソン分布は一定の時間や空間内で起こる「稀な事象」の回数を予測するためのツールです。例えば1時間に平均10回電話が鳴るコールセンターや、1平方メートルあたりに平均5個の花が咲く花畑など具体的な状況を想像してみてください。

ここで、「稀な事象」とは、一定の時間や空間で起こる可能性はあるものの、それが実際に起こる確率は低いという事象を指します。電話が鳴る回数や花が咲く個数など、これらの事象はランダムに発生します。<br>

ポアソン分布の確率質量関数は以下のように表されます。<br>
$P(X=k) = \frac{\lambda^k e^{-\lambda}}{k!}$<br>

ここで、
- $k$ は観察された事象の回数（例：電話が鳴った回数、花が咲いた個数など）
- $e$ は自然対数の底（約2.71828の値を持つ数学的な定数）
- $λ$ は一定の時間や空間で事象が発生する平均回数
- $k!$ は$k$の階乗（$k$から1までの全ての整数の積）を表します。

この式を使うと、特定の時間や空間で事象が $k$ 回発生する確率を計算することができます。このように、ポアソン分布は稀な事象の発生確率を予測する強力なツールとなります。

## ランダムグラフの生成とポアソン分布との比較
ランダムグラフの生成とポアソン分布との比較はネットワークの構造を理解し、その特性を定量化するための一般的な手法です。具体的には、以下のような観点から有用です。<br>

1. **分布** <br>
・ ランダムグラフの**次数分布**（各ノードが持つエッジの数の分布）は理論的に**ポアソン分布**に従うとされています。したがって、実際のネットワークの次数分布をポアソン分布と比較することでネットワークがランダムな接続性を持つかどうかを評価できます。<br>
<br>
2. **ネットワークのランダム性** <br>
・ ポアソン分布との比較により、ネットワークがランダムに生成されたものか、それとも何らかの規則に基づいて生成されたものかを判断することができます。例えば、次数分布がポアソン分布から大きく逸脱している場合、ネットワークは**スケールフリー**な性質を持つ可能性があります。<br>
<br>
3. **モデルの適合性** <br>
・ ポアソン分布との比較はネットワークのモデル化において、どの程度現実のネットワークを反映しているかを評価するための指標となります。ポアソン分布によく適合する場合、ネットワークはエルデシュ・レイニーのランダムグラフモデルによって表現できると言えます。<br>

## スクリプトの説明
このスクリプトは、Erdős-Rényiグラフ（ランダムグラフ）とポアソン分布の比較を目的としています。具体的には、異なる確率 $p$ で生成されたErdős-Rényiグラフの次数分布を、対応するポアソン分布と比較することで、ランダムグラフの次数分布がポアソン分布に従うことを示しています。<br>

1. **異なる生成確率のランダムグラフの生成**<br>
・ スクリプトは異なる確率でエルドシュ・レニー (Erdős-Rényi) ランダムグラフを生成します。<br>
・ nx.erdos_renyi_graph(n, p) 関数は、ノード数 𝑛 とエッジの生成確率 𝑝 をパラメータとして受け取り、エッジが存在する確率が 𝑝 のランダムグラフを生成します。<br>
・ er01 = nx.erdos_renyi_graph(100, 0.01) は、100個のノードを持ち、各エッジが存在する確率が0.01のエルドシュ・レニー・グラフを生成します。<br>
<br>
2. **グラフの描画と次数分布のプロット**<br>
・ 生成したランダムグラフの描画とその次数分布のプロットを行います。<br>
・ nx.draw と plt.plot(nx.degree_histogram(graph)) を使用してグラフの描画と次数分布の可視化を行います。<br>
・ 次数分布はグラフ内の各ノードの次数（接続数）の分布を示し、ランダムグラフの構造を理解するのに役立ちます。<br>
<br>
3. **ポアソン分布との比較**<br>
・ スクリプトの後半ではポアソン分布を生成し、ランダムグラフの次数分布と比較します。<br>
・ ポアソン分布の関数 p1(x), p3(x), p5(x), p10(x) は、それぞれ平均 𝜆 = 1,3,5,10 のポアソン分布を生成します。<br>
・ ポアソン分布とランダムグラフの次数分布の比較により、ランダムグラフのネットワーク特性とポアソン分布の理論的な分布との類似性や相違点を調べることができます。

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import scipy.special

# ポアソン分布を定義
def p1(x):
    return (1**x)*(np.e**(-1))/scipy.special.factorial(x)

def p3(x):
    return (3**x)*(np.e**(-3))/scipy.special.factorial(x)

def p5(x): 
    return (5**x)*(np.e**(-5))/scipy.special.factorial(x)

def p10(x):
    return (10**x)*(np.e**(-10))/scipy.special.factorial(x)

# 確率でErdős-Rényiグラフを生成（ノード数は50に設定していますが適宜調整してください）
er01 = nx.erdos_renyi_graph(50, 0.01)
plt.subplot(3, 4, 1)
nx.draw(er01, node_size=10, node_color='red')
plt.title("ER (p=0.01)")

plt.subplot(3, 4, 5)
plt.plot(nx.degree_histogram(er01))
plt.title("Degree Histogram")

er1 = nx.erdos_renyi_graph(50, 0.03)
plt.subplot(3, 4, 2)
nx.draw(er1, node_size=10, node_color='red')
plt.title("ER (p=0.03)")

plt.subplot(3, 4, 6)
plt.plot(nx.degree_histogram(er1))

er5 = nx.erdos_renyi_graph(50, 0.05)
plt.subplot(3, 4, 3)
nx.draw(er5, node_size=10, node_color='red')
plt.title("ER (p=0.05)")

plt.subplot(3, 4, 7)
plt.plot(nx.degree_histogram(er5))

er8 = nx.erdos_renyi_graph(50, 0.1)
plt.subplot(3, 4, 4)
nx.draw(er8, node_size=10, node_color='red')
plt.title("ER (p=0.1)")

plt.subplot(3, 4, 8)
plt.plot(nx.degree_histogram(er8))

# ポアソン分布のプロット
plt.subplot(3, 4, 9)
x = np.arange(0, 4, 1)
y = p1(x)
plt.plot(x, y)
plt.title("Poisson (λ=1)")

plt.subplot(3, 4, 10)
x = np.arange(0, 10, 1)
y = p3(x)
plt.plot(x, y)
plt.title("Poisson (λ=3)")

plt.subplot(3, 4, 11)
x = np.arange(0, 12, 1)
y = p5(x)
plt.plot(x, y)
plt.title("Poisson (λ=5)")

plt.subplot(3, 4, 12)
x = np.arange(0, 20, 1)
y = p10(x)
plt.plot(x, y)
plt.title("Poisson (λ=10)")

plt.tight_layout()
plt.show()
```

## コンフィグレーションモデル
コンフィグレーションモデルとは、特定の次数分布（各ノードが持つエッジの数の分布）を持つランダムグラフを生成するためのモデルです。このモデルは、現実のネットワークの次数分布を再現するために広く使用されます。<br>

例えば、学校のクラスの友達関係を考えてみましょう。クラスには友達が多い生徒もいれば、少ない生徒もいます。この状況を再現するために、コンフィグレーションモデルを使うと、各生徒（ノード）の友達の数（次数）を指定し、その通りに友達関係（エッジ）をランダムに形成することができます。<br>

生成方法の流れ<br>
**次数の設定：** まず、各ノードの次数（友達の数）を指定します。例えば、ノードAが3本のエッジ、ノードBが2本のエッジを持つとします。<br>

**エッジの割り当て：** エッジ割り当ての過程で自己ループ（同じノードに戻るエッジ）や多重エッジ（同じノード間の複数のエッジ）が発生することがあります。友達関係で考えると、自己ループは「自分自身を友達としてカウントする」ような状況に相当します。このような自己ループや多重エッジは現実のネットワークでは不自然であるため、これらを調整することで、より現実的なネットワークを生成します。<br>

**自己ループと多重エッジの調整：** エッジ割り当ての過程で、自己ループ（同じノードに戻るエッジ）や多重エッジ（同じノード間の複数のエッジ）が発生することがあります。これらを調整することで、より現実的なネットワークを生成します。<br>

**コンフィグレーションモデルの計算**<br>
具体的な計算例として、10個のノードがあり、それぞれのノードの次数が以下の通りであるとします。<br>

ノード1: 3<br>
ノード2: 2<br>
ノード3: 1<br>
・・・（以下省略）<br>
まず、各ノードに指定された次数分の「半エッジ」を与え、それらをランダムに結びつけて完全なエッジを形成します。この過程を繰り返していくことで、指定された次数分布を持つランダムグラフが完成します。<br>

**コンフィグレーションモデルの応用例**<br>
**ソーシャルネットワークの解析：** 現実のソーシャルネットワークの構造を再現し、友達関係のパターンを分析するために使用されます。<br>
**インフラネットワークの設計：** 電力網や通信ネットワークなどのインフラ設計において、特定の次数分布を持つネットワークをシミュレートし、効率的な設計を行うために役立ちます。<br>
**生物学的ネットワークの研究：** 遺伝子ネットワークや神経ネットワークの構造を再現し、それらの機能や特性を理解するために使用されます。<br>

このスクリプトの目的は、設定モデル**Configuration Model**を使用して、特定の次数列（degree sequence）に基づいて3つの異なるネットワークを生成し、それぞれのグラフを可視化することです。<br>

**次数列の定義**<br>
deg_seq = [0, 1, 1, 2, 2, 2, 2, 3, 3, 4]<br>
　・この次数列は、グラフの各ノードの次数を示しています。例えば、次数が4のノードが1つ、次数が3のノードが2つ、次数が2のノードが4つなどを含みます。<br>
 
**設定モデルを使用してグラフを生成**<br>
　・3つの異なるグラフを設定モデルに基づいて生成します。<br>
　・設定モデルは、指定された次数列に従ってランダムにエッジを接続することによってグラフを構築します。<br>

**スプリングレイアウトでグラフを描画**<br>
　・各グラフをスプリングレイアウトを使用して描画します。スプリングレイアウトは、力学モデルに基づいてノードを配置し、グラフの構造を視覚的に理解しやすくします。<br>

**生成される3つのネットワークについて**<br>
**1. ネットワークの特性**<br>
・ 各ネットワークは、同じ次数列に基づいて生成されるため、ノードの次数分布は同じです。しかし、設定モデルはランダムにエッジを接続するため、生成される各ネットワークの具体的な構造は異なります。<br>
・ 設定モデルでは、同じ次数列に対して複数の異なるグラフ構造が生成されることがあります。これにより、同じ次数列でも異なるトポロジーを持つネットワークが得られることを示しています。<br>
**2. 視覚的比較**
・ スクリプトは、同じ次数列に基づいて生成された3つの異なるネットワークを並べて表示します。これにより、設定モデルがどのようにランダムにエッジを接続し、異なるグラフ構造を生成するかを視覚的に比較できます。<br>

**スクリプトの目的**<br>
このスクリプトの主な目的は、設定モデルを使用して、特定の次数列に基づくネットワークの多様性を示すことです。<br>
**1. 次数列の制約を満たす多様なネットワーク構造**<br>
 ・ 同じ次数列でも、ランダムにエッジを接続することによって多様なネットワーク構造が生成されることを示します。<br>
**2. 設定モデルの理解**<br>
・ 設定モデルがどのように機能し、指定された次数列に基づいてグラフを構築するかを理解するための例として使用できます。<br>

```python
import networkx as nx
import matplotlib.pyplot as plt

# ノードの次数列を定義
deg_seq = [0, 1, 1, 2, 2, 2, 2, 3, 3, 4]

# 1つ目のプロットを作成（全体を3つの部分に分割した1つ目の部分）
plt.subplot(1, 3, 1)
# 設定モデルを使用してグラフを生成
G1 = nx.configuration_model(deg_seq)
# スプリングレイアウトでグラフを描画
nx.draw_spring(G1, node_size=10, node_color='red')

# 2つ目のプロットを作成（全体を3つの部分に分割した2つ目の部分）
plt.subplot(1, 3, 2)
# 設定モデルを使用してグラフを生成
G2 = nx.configuration_model(deg_seq)
# スプリングレイアウトでグラフを描画
nx.draw_spring(G2, node_size=10, node_color='red')

# 3つ目のプロットを作成（全体を3つの部分に分割した3つ目の部分）
plt.subplot(1, 3, 3)
# 設定モデルを使用してグラフを生成
G3 = nx.configuration_model(deg_seq)
# スプリングレイアウトでグラフを描画
nx.draw_spring(G3, node_size=10, node_color='red')

# 全てのプロットを表示
plt.show()
```

## スケールフリーネットワーク
**スケールフリーネットワーク**はノード同士の結びつきの数（次数）が特定のパターンに従うネットワークであり、次数が非常に多い<span style="color: red;">**ハブ**</span>と呼ばれるノードが存在する特徴を持ちます。一般的に、スケールフリーネットワークは**次数分布がべき乗則に従う**傾向があります。<br>

例えば、SNSでは、一部のユーザーが多くのフォロワーを持ち、他のユーザーよりも影響力があります。これらのユーザーはインフルエンサーと呼ばれ、ネットワーク内でのハブとして機能します。その結果、**フォロワー数の分布はべき乗則に従い、少数のユーザーが非常に多くのフォロワーを持つ一方で、多くのユーザーは比較的少数のフォロワーしか持ちません。** 情報の拡散も、これらのインフルエンサーやフォロワーの多いユーザーが中心となり、SNS上での投稿が迅速に多くのユーザーに広まる特性を持ちます。<br>
<br>

スケールフリーネットワークの解析は、さまざまな場面で役立ちます。<br>
1. **情報の拡散とマーケティング戦略**<br>
・ スケールフリーネットワークでは、少数のインフルエンサーやハブが情報の中心となります。したがって、特定の製品やサービスの宣伝やマーケティング戦略を設計する際には、これらのハブに焦点を当てることが効果的です。彼らを通じて広告を行うことで、情報が効率的に多くのユーザーに届けられる可能性が高まります。<br>

2. **疾病の拡散の予測と対策**<br>
・ エピデミックモデリングにおいて、スケールフリーネットワークの解析は重要な役割を果たします。特定の感染症がハブを介してどのように広がるかを理解し、それに基づいて効果的な予防策や制御策を導くことが可能です。ハブを重点的に監視し、早期に介入することで、感染の拡大を抑えることができる場合があります。<br>

3. **交通ネットワークや物流の最適化**<br>
・ 輸送や物流におけるネットワーク解析では、スケールフリーネットワークの理解が必要です。特定の交通ハブや物流センターがどのようにして物資や情報を効率的に配送するかを分析し、その最適化に役立てることができます。<br>

4. **ソーシャルネットワークの分析と影響力の測定**<br>
・ SNSやその他のソーシャルネットワークにおいて、特定のユーザーがどの程度影響力を持ち、他のユーザーに与える影響の大きさを理解するために、スケールフリーネットワークの解析が有効です。この情報は、社会学やマーケティングにおいて重要なインサイトを提供します。<br>

これらの例からもわかるように、スケールフリーネットワークの理解と解析は、現代社会でのさまざまなネットワーク現象の理解と対策に不可欠です。

以下のスクリプトでスケールフリーネットワークの特徴である次数分布がべき乗則に従うことを確認できます。この分布は、少数のハブが多くのリンクを持ち、多数のノードが比較的少ないリンクしか持たないという現象を示します。

次数が最大のノードのみを赤色で表記しています。

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# スケールフリーネットワークの生成
ba_graph = nx.barabasi_albert_graph(100, 5)

# 最も多くのエッジを持つノードを特定する
max_degree_node = max(ba_graph.nodes(), key=lambda n: ba_graph.degree(n))

# ノード数、エッジ数、平均次数の計算
num_nodes = nx.number_of_nodes(ba_graph)
num_edges = nx.number_of_edges(ba_graph)
average_degree = 2 * num_edges / num_nodes

# 計算結果をコンソールに出力
print("Number of nodes:", num_nodes)
print("Number of edges:", num_edges)
print("Average degree:", '{:.3f}'.format(average_degree))

# ネットワークの可視化
plt.figure(figsize=(16, 6))

# ノードの色を定義するためのリスト
node_colors = ['red' if n == max_degree_node else 'blue' for n in ba_graph.nodes()]

plt.subplot(131)
nx.draw(ba_graph, with_labels=False, node_color=node_colors, node_size=30, edge_color='gray')
plt.title('Scale-Free Network with Node with Maximum Degree Highlighted')

# 次数分布の計算とプロット
degree_sequence = sorted([d for n, d in ba_graph.degree()], reverse=True)

plt.subplot(132)
plt.hist(degree_sequence, bins=20, alpha=0.7)
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.title('Degree Distribution of Scale-Free Network')

plt.subplot(133)
plt.hist(degree_sequence, bins=20, alpha=0.7, log=True)
plt.xlabel('Degree (log scale)')
plt.ylabel('Frequency')
plt.title('Degree Distribution (Log Scale)')

plt.tight_layout()
plt.show()
```

## スモールワールド
**スモールワールドグラフは平均最短経路が比較的に短く、ノード間の接続が少ないホップで行われることを特徴とします。同時に高いクラスタリング係数を示し、ノードの近傍が密に結びついている領域（クラスター）が存在します。** この特性はランダムグラフと規則的なグラフの中間に位置することを意味し、Watts-Strogatzモデルなどを用いて生成されます。

スモールワールドグラフはネットワーク内の個々の要素が直接的につながっており、同時にクラスターを形成しているという点で、**実世界の友人や同僚のグループと似ています。** 例えば、友人の友人があなたに直接つながっていることが多い社会ネットワークでは、短い経路で広がる情報や影響が特徴です。同時に、一部の中心的な人物が多くの人々と直接つながっているため、情報が効率的に広まる構造も備えています。

<span style="color: red;">**スケールフリーネットワークとスモールワールドネットワークの主な違いは次数分布にあります。**</span> スケールフリーなネットワークでは、結節（ノード）の次数分布がべき乗則に従い、ほとんどのノードが少数の接続しか持たず、一部の**ハブ**（高次数のノード）が非常に多くの接続を持っています。
一方、スモールワールドネットワークでは**次数分布に特に規則はなく、ランダム性と規則性の中間に位置する**ため、一様ではない分布を示します。

### スケールフリーネットワークとスモールワールドネットワークの違い
1. **結節の次数分布**<br>
・ **スケールフリー：** 結節（ノード）の次数分布が**べき乗則**（べき分布）に従います。つまり、ほとんどのノードが少数の接続しか持たず、一部の**ハブ**（高次数のノード）が非常に多くの接続を持っています。<br>
・ **スモールワールド：** 次数分布に**特定の規則はなく**、ランダム性と規則性の中間に位置するグラフです。次数分布が特に特定のパターンに従う必要はありません。<br>

2. **ネットワーク構造の生成メカニズム**<br>
・**スケールフリー：** 結節の新規追加時に既存のノードに接続する確率がその次数に比例する**優先的選択**メカニズムに基づいて生成されます（例: バラバシ-アルベルトモデル）<br>
・ **スモールワールド：** 初期の規則的な構造を**ランダムに再接続**することで生成されます（例: Watts-Strogatzモデル）<br>

3. **情報伝播の特性**<br>
・ **スケールフリー：** **ハブノードによって情報が効率的に伝播**し、ネットワーク全体での情報の影響力が大きい特性があります。<br>
・ **スモールワールド：** 短い経路で情報が拡散しやすく、**クラスター化された構造によって情報が集中して伝わる**傾向があります。<br>

```python
import networkx as nx
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))

# 初期化
plt.subplot(1, 3, 1)
ws0 = nx.watts_strogatz_graph(100, 4, 0)
nx.draw_circular(ws0, node_size=10, node_color='red')
print("rewiring 0%")
print("L =", nx.average_shortest_path_length(ws0))
print("C =", nx.average_clustering(ws0))

# 30% リワイアリング
plt.subplot(1, 3, 2)
ws03 = nx.watts_strogatz_graph(100, 4, 0.3)
nx.draw_circular(ws03, node_size=10, node_color='red')
print("rewiring 30%")
print("L =", nx.average_shortest_path_length(ws03))
print("C =", nx.average_clustering(ws03))

# 100% リワイアリング
plt.subplot(1, 3, 3)
ws1 = nx.watts_strogatz_graph(100, 4, 1)
nx.draw_circular(ws1, node_size=10, node_color='red')
print("rewiring 100%")
print("L =", nx.average_shortest_path_length(ws1))
print("C =", nx.average_clustering(ws1))

plt.show()
```

## 一休み･･･
GN（Girvan-Newman）ベンチマークは、Girvan-Newmanアルゴリズムとは異なるものです。<br>
**Girvan-Newmanアルゴリズム**<br>
これは、ネットワークのコミュニティ検出を行うためのアルゴリズムです。このアルゴリズムは、エッジの媒介中心性（betweenness centrality）を使用してネットワークの構造を解析し、その結果をもとにコミュニティ（モジュール）を特定します。エッジの媒介中心性は、ネットワーク内の各エッジが異なるノード対間の最短経路にどれだけ貢献しているかを示す指標です。<br>

**GN（Girvan-Newman）ベンチマーク**<br>
これは、ネットワーク科学やコンピュータサイエンスにおいて、ネットワークアルゴリズムの性能評価を目的としたベンチマークです。GNベンチマークは、特定の性質（例えば、コミュニティ構造の知識）を持つ人工的なネットワークを生成することによって、アルゴリズムの効率性や精度をテストするために使用されます。このベンチマークの名前は、GirvanとNewmanにちなんで名付けられていますが、具体的なアルゴリズムとは直接的に関連しているわけではありません。<br>

要するに、Girvan-Newmanアルゴリズムは特定のネットワーク解析手法を指し、一方でGNベンチマークはアルゴリズムの評価や比較のための人工的なネットワーク生成手法を指します。<br>



## LFRベンチマークグラフの生成と可視化
LFR（Lancichinetti-Fortunato-Radicchi）ベンチマークは、LFRは次数分布とコミュニティサイズ分布が裾の重い（パワー・ロー）であるネットワークを生成します。この特性により、実世界の多くのネットワークに見られるハブノード（次数が非常に大きいノード）や大きなコミュニティが反映されやすくなります。
LFRベンチマークはミキシングパラメータを用いてネットワーク内のコミュニティ間のエッジの存在確率を調整します。このパラメータの調整により、生成されるネットワークのモジュラリティ（コミュニティの強度）を制御することができます。
また、GN（Girvan-Newman）ベンチマークと比較して、LFRベンチマークはより現実のネットワーク特性に近いものを生成します。特に、次数分布の裾の重さやコミュニティのサイズ分布の非均一性などが実データによく類似しています。
LFRベンチマークは、新しいネットワーク解析手法やコミュニティ検出アルゴリズムの評価に広く使用されています。生成されたネットワークの特性が既知の実データセットとどれだけ類似しているかを評価することで、モデルの有用性や適用可能性を評価することができます。<br>
<br>

このスクリプトを実行することで、LFRベンチマークグラフの生成とそのネットワーク構造の視覚的理解を深めることができます。<br>

**1. ベンチマークグラフの生成**<br>
LFR_benchmark_graph 関数を使用して、指定されたパラメータ（ノード数、次数分布の指数、コミュニティサイズ分布の指数、ミキシングパラメータなど）に基づいてLFRベンチマークグラフを生成します。このグラフは、実世界の複雑なネットワークを模倣するために使用されます。<br>

**2. コミュニティの抽出**<br>
生成したネットワークから各ノードのコミュニティを抽出し、整数のリストとして表現します。各ノードは、一意のコミュニティに所属しています。<br>

**3. ノードのサイズ設定**<br>
各ノードのサイズは、そのノードの次数に比例して設定されます。このスクリプトでは、サイズの倍率を調整可能です（例えば、10 の倍率が使用されています）。<br>

**4. 描画**<br>
ノードの位置はSpringレイアウトを用いて設定され、各ノードの色は所属するコミュニティに基づいて色分けされます。カラーマップは plt.cm.jet が使用されています。<br>

```python
import networkx as nx
from networkx.generators.community import LFR_benchmark_graph
import matplotlib.pyplot as plt

n = 128  # ノード数を128に変更
tau1 = 3  # ノードの次数分布の指数
tau2 = 1.5  # コミュニティサイズ分布の指数
mu = 0.1  # ミキシングパラメータ
# min_communityとmax_communityを調整
min_community = 10
max_community = 20

G = LFR_benchmark_graph(n, tau1, tau2, mu, average_degree=5, min_community=min_community, max_community=max_community, seed=10)

# 各ノードのコミュニティを整数のリストに変換
communities = {n: list(d['community'])[0] for n, d in G.nodes(data=True)}

# ノードの次数に基づいてノードサイズを設定
node_sizes = [G.degree(n) * 10 for n in G.nodes()]  # 10は倍率です。適宜調整してください。

pos = nx.spring_layout(G)
nx.draw(G, pos, node_size=node_sizes, node_color=[communities[n] for n in G.nodes()], cmap=plt.cm.jet)
plt.show()
```

# たとえば空手クラブのデータを使って解析
## 注). 解析については８章で触れますが、ここでは手順を理解することが主な目的です
**1. 次数分布の取得・比較:** 空手クラブの実データの次数分布を取得し、比較するためのモデル（例えばBAモデルなど）と比較します。これにより、ネットワークの特性やスケールフリー性を理解します。

**2. 次数分布の結果から中心性を求める方向性を決定:** 次数分布の特性（平均次数、最大次数など）を基に、どの中心性指標を計算するかの方向性を決定します。例えば、高次数中心性を持つノードを特定することができます。

**3. 抽出または分割:** コミュニティ検出アルゴリズムを使用してネットワークを適切なコミュニティに抽出または分割します。これにより、ネットワーク内の構造や重要なサブグループを特定します。

**4. グラフの可視化:** 分割されたネットワークや抽出されたコミュニティを色分けしてグラフとして可視化します。これにより、ネットワーク全体や各コミュニティの構造や関係を視覚的に理解できます。

**5. 定量的評価:** 分割後のコミュニティの質やモデルとの適合度を定量的に評価します。モジュラリティなどの指標を使用して、分割されたコミュニティがどれだけ自己結合性を持っているかを確認します。

## 空手クラブグラフとバラバシ-アルバートモデルの次数分布比較とコルモゴロフ-スミルノフ検定

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import collections
import scipy.stats as stats

# 空手クラブのグラフを読み込む
G = nx.karate_club_graph()

# ノード数とエッジ数を取得
n = G.number_of_nodes()
m = G.number_of_edges()

# 空手クラブのグラフの次数分布を取得
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
degree_count = collections.Counter(degree_sequence)
deg, cnt = zip(*degree_count.items())

# スケールフリー特性を確認するために次数分布をプロット
plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.bar(deg, cnt, width=0.80, color='b')
plt.title("Degree Histogram (Karate Club)")
plt.ylabel("Count")
plt.xlabel("Degree")

plt.subplot(122)
plt.loglog(deg, cnt, marker='o', linestyle='None')
plt.title("Log-Log Degree Distribution (Karate Club)")
plt.ylabel("Frequency")
plt.xlabel("Degree")

plt.tight_layout()
plt.show()

# Barabási-Albertモデルを使用してスケールフリーグラフを生成
# パラメータmは空手クラブの平均次数に基づいて設定
mean_degree = np.mean(degree_sequence)
m = int(mean_degree / 2)  # BAモデルのパラメータmは各新しいノードが接続する既存ノードの数

G_ba = nx.barabasi_albert_graph(n, m)

# BAモデルの次数分布を取得
degree_sequence_ba = sorted([d for n, d in G_ba.degree()], reverse=True)
degree_count_ba = collections.Counter(degree_sequence_ba)
deg_ba, cnt_ba = zip(*degree_count_ba.items())

# BAモデルと空手クラブのグラフの比較のために次数分布をプロット
plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.bar(deg_ba, cnt_ba, width=0.80, color='g')
plt.title("Degree Histogram (BA Model)")
plt.ylabel("Count")
plt.xlabel("Degree")

plt.subplot(122)
plt.loglog(deg_ba, cnt_ba, marker='o', linestyle='None')
plt.title("Log-Log Degree Distribution (BA Model)")
plt.ylabel("Frequency")
plt.xlabel("Degree")

plt.tight_layout()
plt.show()

# 空手クラブのグラフとBAモデルのグラフの可視化
plt.figure(figsize=(12, 6))

plt.subplot(121)
nx.draw_spring(G, with_labels=True, node_color='skyblue', edge_color='gray', node_size=500, font_size=10)
plt.title("Original Karate Club Graph")

plt.subplot(122)
nx.draw_spring(G_ba, with_labels=True, node_color='lightgreen', edge_color='gray', node_size=500, font_size=10)
plt.title("BA Model Graph")

plt.tight_layout()
plt.show()

# 定量的評価
mean_degree_ba = np.mean(degree_sequence_ba)
max_degree_karate = max(degree_sequence)
max_degree_ba = max(degree_sequence_ba)

print("Karate Club Graph:")
print(f"Mean Degree: {mean_degree:.2f}")
print(f"Max Degree: {max_degree_karate}")

# 空手クラブの最大次数に接続するノード数を表示
for node in G.nodes():
    if G.degree(node) == max_degree_karate:
        neighbors_count = len(list(G.neighbors(node)))
        print(f"Nodes with max degree ({max_degree_karate}) in the Karate Club graph are connected to {neighbors_count} neighbors.")

print("\nBA Model Graph:")
print(f"Mean Degree: {mean_degree_ba:.2f}")
print(f"Max Degree: {max_degree_ba}")

# 統計的検定 (コルモゴロフ-スミルノフ検定)
ks_stat, p_value = stats.ks_2samp(degree_sequence, degree_sequence_ba)

print("\nKolmogorov-Smirnov Test:")
print(f"KS Statistic: {ks_stat:.2f}")
print(f"P-value: {p_value:.2f}")

# 検定結果の解釈
if p_value > 0.05:
    print("The degree distributions are not significantly different (fail to reject H0).")
else:
    print("The degree distributions are significantly different (reject H0).")
```

**KS検定を実行する意味**<br>
KS検定は、2つの度数分布が統計的に有意に異なるかどうかを検証するために使用されます。この場合、Karate Club GraphとBA Model Graphの次数分布を比較しています。

意義と目的<br>
モデルの妥当性の評価:
Karate Club Graphは実際のデータに基づくグラフです。一方、BA Model Graphはバラバシ-アルバートモデルに基づく生成されたランダムグラフです。
KS検定を使用してこれら2つのグラフの次数分布を比較することで、BAモデルがどれだけ現実のデータ（Karate Club Graph）を再現しているかを評価できます。<br>

統計的な検証:<br>
平均次数や最大次数といった基本統計量だけでは、分布全体の形状の違いを捉えるのに限界があります。KS検定は累積分布関数の差異を評価するため、より詳細な比較が可能です。<br>

帰無仮説の検証:<br>

KS検定の結果、p値が0.86と非常に高く、帰無仮説を棄却することができません。つまり、Karate Club GraphとBA Model Graphの次数分布は統計的に有意に異ならないということです。<br>

## 空手クラブグラフの解析とコミュニティ検出
このスクリプトでは、空手クラブのグラフを読み込み、ノードとエッジ数を取得し、次数分布をプロットして中心性を求め、Louvain法によるコミュニティ検出を行い、最後にオリジナルグラフとBAモデルで生成されたグラフを比較します。

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import community
import numpy as np
import community as community_louvain

# 1. 空手クラブのグラフを読み込み、オリジナルグラフとして描画します。
G = nx.karate_club_graph()

# 2. ノードとエッジ数を取得します。
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()

# 上位3位までのエッジ数を持つノードの情報をコンソールに出力します。
node_degrees = dict(G.degree())
sorted_node_degrees = sorted(node_degrees.items(), key=lambda x: x[1], reverse=True)[:3]
print(f"Number of nodes: {num_nodes}")
print(f"Number of edges: {num_edges}")
print("Top 3 nodes with highest degrees:")
for node, degree in sorted_node_degrees:
    print(f"Node {node}: {degree} edges")

# 3. 次数分布の取得
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
degree_count = np.bincount(degree_sequence)
degrees = np.arange(len(degree_count))

# 4. 次数中心性を求め、得られた結果をノードに表示します。
degree_centrality = nx.degree_centrality(G)
nx.set_node_attributes(G, degree_centrality, 'degree_centrality')

# 5. Louvain法を用いてコミュニティ抽出を行います。
partition = community_louvain.best_partition(G)
community_colors = [partition[node] for node in G]

# 6. モジュラリティなどの指標を使用して、分割されたコミュニティがどれだけ自己結合性を持っているかを確認します。
modularity = community_louvain.modularity(partition, G)
print(f"Modularity: {modularity}")

# 7. BAモデルでスケールフリーグラフを生成し可視化。
BA_G = nx.barabasi_albert_graph(num_nodes, 2)

# グラフの可視化
plt.figure(figsize=(16, 16))

# オリジナルグラフ
plt.subplot(321)
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10)
plt.title('Original Karate Club Graph')

# BAモデルグラフ
plt.subplot(322)
nx.draw(BA_G, with_labels=True, node_color='lightgreen', edge_color='gray', node_size=500, font_size=10)
plt.title('Barabasi-Albert Model Graph')

# 次数分布のヒストグラム
plt.subplot(323)
plt.bar(degrees, degree_count, color='b')
plt.title("Degree Distribution of Karate Club Graph")
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.yscale('log')
plt.xscale('log')

# 次数中心性の可視化
plt.subplot(324)
pos = nx.spring_layout(G)
node_color = [degree_centrality[node] for node in G]
nx.draw(G, pos, with_labels=True, node_color=node_color, node_size=500, cmap=plt.cm.Blues)
plt.title("Degree Centrality of Karate Club Graph")

# 次数分布のヒストグラム
plt.subplot(325)
plt.hist(degree_sequence, bins=range(min(degree_sequence), max(degree_sequence) + 1), color='b', alpha=0.7)
plt.title("Degree Histogram")
plt.xlabel("Degree")
plt.ylabel("Frequency")

# 次数中心性のヒストグラム
plt.subplot(326)
plt.hist(degree_centrality.values(), bins=20, color='b', alpha=0.7)
plt.title("Degree Centrality Histogram")
plt.xlabel("Degree Centrality")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

# Louvainコミュニティ検出の可視化
plt.figure(figsize=(8, 8))
nx.draw(G, pos, node_color=community_colors, with_labels=True, cmap=plt.cm.rainbow, node_size=500)
plt.title("Louvain Community Detection")
plt.show()
```

## 空手クラブグラフの解析とGirvan-Newmanコミュニティ検出
Girvan-Newmanアルゴリズムを用いて**コミュニティの数が10以下になるまで反復的にコミュニティを抽出します。** その後、モジュラリティを計算しています。

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import community
import numpy as np
import itertools

# 1. 空手クラブのグラフを読み込み、オリジナルグラフとして描画します。
G = nx.karate_club_graph()

# 2. ノードとエッジ数を取得します。
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()

# 上位3位までのエッジ数を持つノードの情報をコンソールに出力します。
node_degrees = dict(G.degree())
sorted_node_degrees = sorted(node_degrees.items(), key=lambda x: x[1], reverse=True)[:3]
print(f"Number of nodes: {num_nodes}")
print(f"Number of edges: {num_edges}")
print("Top 3 nodes with highest degrees:")
for node, degree in sorted_node_degrees:
    print(f"Node {node}: {degree} edges")

# 3. 次数分布の取得
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
degree_count = np.bincount(degree_sequence)
degrees = np.arange(len(degree_count))

# 4. 媒介中心性を求め、得られた結果をノードに表示します。
betweenness_centrality = nx.betweenness_centrality(G)
nx.set_node_attributes(G, betweenness_centrality, 'betweenness_centrality')

# 5. Girvan-Newmanアルゴリズムを用いてコミュニティ抽出を行います。
comp = community.girvan_newman(G)
limited = itertools.takewhile(lambda c: len(c) <= 10, comp)
communities = list(limited)[-1]
community_map = {}
for i, com in enumerate(communities):
    for node in com:
        community_map[node] = i
community_colors = [community_map[node] for node in G]

# 6. モジュラリティなどの指標を使用して、分割されたコミュニティがどれだけ自己結合性を持っているかを確認します。
modularity = community.modularity(G, communities)
print(f"Modularity: {modularity}")

# 7. BAモデルでスケールフリーグラフを生成し可視化。
BA_G = nx.barabasi_albert_graph(num_nodes, 2)

# グラフの可視化
plt.figure(figsize=(16, 16))

# オリジナルグラフ
plt.subplot(321)
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10)
plt.title('Original Karate Club Graph')

# BAモデルグラフ
plt.subplot(322)
nx.draw(BA_G, with_labels=True, node_color='lightgreen', edge_color='gray', node_size=500, font_size=10)
plt.title('Barabasi-Albert Model Graph')

# 次数分布のヒストグラム
plt.subplot(323)
plt.bar(degrees, degree_count, color='b')
plt.title("Degree Distribution of Karate Club Graph")
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.yscale('log')
plt.xscale('log')

# 媒介中心性の可視化
plt.subplot(324)
pos = nx.spring_layout(G)
node_color = [betweenness_centrality[node] for node in G]
nx.draw(G, pos, with_labels=True, node_color=node_color, node_size=500, cmap=plt.cm.Blues)
plt.title("Betweenness Centrality of Karate Club Graph")

# 次数分布のヒストグラム
plt.subplot(325)
plt.hist(degree_sequence, bins=range(min(degree_sequence), max(degree_sequence) + 1), color='b', alpha=0.7)
plt.title("Degree Histogram")
plt.xlabel("Degree")
plt.ylabel("Frequency")

# 媒介中心性のヒストグラム
plt.subplot(326)
plt.hist(list(betweenness_centrality.values()), bins=20, color='b', alpha=0.7)
plt.title("Betweenness Centrality Histogram")
plt.xlabel("Betweenness Centrality")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

# Girvan-Newmanコミュニティ検出の可視化
plt.figure(figsize=(8, 8))
nx.draw(G, pos, node_color=community_colors, with_labels=True, cmap=plt.cm.rainbow, node_size=500)
plt.title("Girvan-Newman Community Detection")
plt.show()
```

## 空手クラブグラフの解析とGirvan-Newmanコミュニティ検出
Girvan-Newmanアルゴリズムを用いて**コミュニティが2つになるまで反復的にコミュニティを抽出します。** 最終的に2つのコミュニティを得て、それに基づいてモジュラリティを計算しています。


```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import community
import numpy as np
import itertools

# 1. 空手クラブのグラフを読み込み、オリジナルグラフとして描画します。
G = nx.karate_club_graph()

# 2. ノードとエッジ数を取得します。
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()

# 上位3位までのエッジ数を持つノードの情報をコンソールに出力します。
node_degrees = dict(G.degree())
sorted_node_degrees = sorted(node_degrees.items(), key=lambda x: x[1], reverse=True)[:3]
print(f"Number of nodes: {num_nodes}")
print(f"Number of edges: {num_edges}")
print("Top 3 nodes with highest degrees:")
for node, degree in sorted_node_degrees:
    print(f"Node {node}: {degree} edges")

# 3. 次数分布の取得
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
degree_count = np.bincount(degree_sequence)
degrees = np.arange(len(degree_count))

# 4. 媒介中心性を求め、得られた結果をノードに表示します。
betweenness_centrality = nx.betweenness_centrality(G)
nx.set_node_attributes(G, betweenness_centrality, 'betweenness_centrality')

# 5. Girvan-Newmanアルゴリズムを用いてコミュニティ抽出を行います。
comp = community.girvan_newman(G)
# コミュニティが2つになるまで反復処理を行います。
for communities in comp:
    if len(communities) == 2:
        break

community_map = {}
for i, com in enumerate(communities):
    for node in com:
        community_map[node] = i
community_colors = [community_map[node] for node in G]

# 6. モジュラリティなどの指標を使用して、分割されたコミュニティがどれだけ自己結合性を持っているかを確認します。
modularity = community.modularity(G, communities)
print(f"Modularity: {modularity}")

# 7. BAモデルでスケールフリーグラフを生成し可視化。
BA_G = nx.barabasi_albert_graph(num_nodes, 2)

# グラフの可視化
plt.figure(figsize=(16, 16))

# オリジナルグラフ
plt.subplot(321)
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10)
plt.title('Original Karate Club Graph')

# BAモデルグラフ
plt.subplot(322)
nx.draw(BA_G, with_labels=True, node_color='lightgreen', edge_color='gray', node_size=500, font_size=10)
plt.title('Barabasi-Albert Model Graph')

# 次数分布のヒストグラム
plt.subplot(323)
plt.bar(degrees, degree_count, color='b')
plt.title("Degree Distribution of Karate Club Graph")
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.yscale('log')
plt.xscale('log')

# 媒介中心性の可視化
plt.subplot(324)
pos = nx.spring_layout(G)
node_color = [betweenness_centrality[node] for node in G]
nx.draw(G, pos, with_labels=True, node_color=node_color, node_size=500, cmap=plt.cm.Blues)
plt.title("Betweenness Centrality of Karate Club Graph")

# 次数分布のヒストグラム
plt.subplot(325)
plt.hist(degree_sequence, bins=range(min(degree_sequence), max(degree_sequence) + 1), color='b', alpha=0.7)
plt.title("Degree Histogram")
plt.xlabel("Degree")
plt.ylabel("Frequency")

# 媒介中心性のヒストグラム
plt.subplot(326)
plt.hist(list(betweenness_centrality.values()), bins=20, color='b', alpha=0.7)
plt.title("Betweenness Centrality Histogram")
plt.xlabel("Betweenness Centrality")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

# Girvan-Newmanコミュニティ検出の可視化
plt.figure(figsize=(8, 8))
nx.draw(G, pos, node_color=community_colors, with_labels=True, cmap=plt.cm.rainbow, node_size=500)
plt.title("Girvan-Newman Community Detection")
plt.show()
```
