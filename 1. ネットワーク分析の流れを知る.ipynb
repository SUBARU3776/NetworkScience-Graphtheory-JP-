# ネットワーク分析
ネットワーク分析とは、**ノード**（頂点）と**エッジ**（線）で構成されるグラフを用いて、複雑な関係性や構造を解析する手法です。これにより、ネットワーク内のパターンや特性を理解し、特定の目的に対する洞察を得ることができます。ネットワーク分析は、社会科学、生物学、情報科学など、多岐にわたる分野で応用されています。以下は、ネットワーク分析の基本概念とその流れについての説明です。<br>

## ネットワークの基本概念
1. **ノード（Nodes）**<br>
    ・**ネットワーク内の個々の点**を指します。例えば、人間関係のネットワークではノードは個人を表します。<br>
<br>
2. **エッジ（Edges）**<br> 
    ・**ノード同士を結ぶ線**を指します。これらは関係性や相互作用を示します。<br>
<br>
3. **ネットワーク（Network）**<br>
    ・**ノード**と**エッジ**の集合からなる**グラフ全体**を指します。<br>
<br>
4. **有向グラフ（Directed Graph）**<br>
    ・エッジに**方向がある**グラフです。例えば、ツイッターのフォロー関係は有向グラフです。<br>
<br>
5. **無向グラフ（Undirected Graph）**<br>
    ・エッジに**方向がない**グラフです。例えば、フェイスブックの友人関係は無向グラフです。<br>


## おおまかなネットワーク分析の流れ
1. **データの収集と前処理**<br>
    ・ネットワークデータの収集<br>
    ・データのクリーニングと整形<br>
    ・適切なデータ構造への変換（隣接行列、隣接リストなど）<br>
2. **グラフの構築**<br>
    ・ノードとエッジの定義<br>
    ・グラフオブジェクトの作成（多くの場合、専用のライブラリを使用）<br>
3. **基本的な特徴量の計算**<br>
    ・ノード数、エッジ数の計算<br>
    ・平均次数、密度の計算<br>
    ・連結成分の特定<br>
4. **中心性指標の計算**<br>
    ・次数中心性<br>
    ・固有ベクトル中心性<br>
    ・近接中心性<br>
    ・媒介中心性<br>
5. **クラスター分析**<br>
    ・クラスタリング係数の計算<br>
    ・コミュニティ検出アルゴリズムの適用<br>
6. **パス関連の分析**<br>
    ・平均経路長の計算<br>
    ・ネットワーク直径の計算<br>
    ・最短経路の特定<br>
7. **次数分布の分析**<br>
    ・次数分布のヒストグラム作成<br>
    ・スケールフリー性の確認（べき乗則の適合度チェック）<br>
8. **ネットワークモデルとの比較**<br>
    ・ランダムグラフモデル<br>
    ・スモールワールドモデル<br>
    ・スケールフリーモデル<br>
    ・実際のネットワークとモデルとの比較<br>
9. **動的ネットワーク分析（時系列データがある場合）**<br>
    ・時間経過に伴うネットワーク構造の変化の分析<br>
    ・ノードやエッジの出現・消滅パターンの分析<br>
10. **可視化**<br>
    ・ネットワーク構造の可視化<br>
    ・特徴量や分析結果の可視化（ヒストグラム、散布図など）<br>
11. **統計的検定と解釈**<br>
    ・必要に応じて統計的検定の実施<br>
    ・結果の解釈と考察<br>
12. **応用と予測**<br>
    ・得られた知見の応用（例：情報伝播の予測、影響力のあるノードの特定）<br>
    ・必要に応じた予測モデルの構築<br>
13. **レポート作成と結果の共有**<br>
    ・分析結果のまとめ<br>
    ・視覚化資料の作成<br>
    ・結果の共有とディスカッション<br>

この手順は、問題の性質や利用可能なデータ、分析の目的によって適宜調整されます。また、多くの場合、これらのステップは反復的に行われ、前のステップに戻って再分析を行うこともあります。

グラフ理論を用いた解析では、Python（NetworkX, igraph）、R（igraph）、MATLAB、Gephi などのツールやライブラリがよく使用されます。これらのツールを使うことで、複雑なアルゴリズムを効率的に実装し、大規模なネットワークデータを処理することが可能になります。

```python
import os

current_directory = os.getcwd()
print("Current directory:", current_directory)

new_directory = r"D:\Python\NetworkX\lesmis"# ご自身の利用環境に応じてパスを変更してください

try:
    os.chdir(new_directory)# カレントディレクトリを新しいパスに変更
    print("Current directory is now:", os.getcwd())# 現在のカレントディレクトリを表示
except OSError:
    print(f"Failed to change directory to {new_directory}")# 変更に失敗した場合のエラーメッセージ
```

## 異なるレイアウトで可視化
このスクリプトは、ネットワークグラフの作成と操作を行うためのライブラリnetworkx、グラフや図を描画するためのライブラリmatplotlib.pyplot、ファイルシステム操作を行うためのライブラリosをインポートすることから始まります。<br>
ネットワークグラフを描画する際に、異なるレイアウト（Circular, Kamada-Kawai, Random, Spectral, Spring, Shell）を使用することには重要な意味があります。**各レイアウトは異なる視点からネットワークの構造や特徴を強調する**ため、複数のレイアウトを試すことで、ネットワークのより深い理解を得ることができます。<br>
- Circularレイアウトはノードを円形に配置し、全体の対称性を視覚化します。
- Kamada-Kawaiレイアウトはエッジの長さを均等にし、ノード間の関係を直感的に示します。
- Randomレイアウトはノードをランダムに配置し、エッジの密度や接続性を観察します。
- Spectralレイアウトはネットワークのスペクトル特性やコミュニティ構造を明確にします。
- Springレイアウトはエネルギーの最小化に基づいてノードを配置し、相互作用を視覚化します。
- Shellレイアウトは同心円状にノードを配置し、階層構造や中心性を強調します。

実際のネットワーク分析の手順として、初めに異なるレイアウトでネットワークグラフを描画することは適切で有益です。これは以下の理由からです。<br>
1. **ネットワークの全体像を把握**： 異なるレイアウトを使ってグラフを描画することで、ネットワークの全体的な構造やパターンを視覚的に把握できます。これにより、重要な特徴や異常な点を初期段階で発見することができます。

2. **多角的な視点の提供**： 各レイアウトはネットワークの異なる側面を強調します。例えば、Circularレイアウトは対称性を示し、Springレイアウトはノード間の距離感を強調します。これにより、ネットワークの構造を多角的に理解することができます。

3. **解析のガイド**： 初期の視覚化により、どの部分に焦点を当てるべきか、どの解析手法を適用すべきかのガイドとなります。例えば、クラスタリングの可能性がある部分や、重要なハブノードが見つかるかもしれません。

4. **コミュニケーション**： 初期段階での視覚化は、同僚やステークホルダーとのコミュニケーションにも役立ちます。視覚的な情報は複雑なネットワークの特徴を理解しやすくし、分析の方向性についての共通認識を形成するのに役立ちます。


可読性が低下しますが、このスクリプトはどのような処理が行われているのかを説明するための補足が多くなってます。

```python
# 必要なライブラリをインポート
import networkx as nx  # ネットワークグラフの作成と操作を行うためのライブラリ
import matplotlib.pyplot as plt  # グラフや図を描画するためのライブラリ
import os  # ファイルシステム操作を行うためのライブラリ

# カレントディレクトリを変更
new_directory = r"D:\Python\NetworkX\lesmis"  # ご自身の環境に応じてディレクトリのパスを指定してください
os.chdir(new_directory)  # 指定したディレクトリにカレントディレクトリを変更

# ファイルパスを指定
file_path = r'lesmis.gml'  # 読み込むGMLファイルのパスを指定

# GML形式のファイルを読み込み
G = nx.readwrite.gml.read_gml(file_path)  # 指定されたGMLファイルを読み込み、ネットワークグラフオブジェクトを作成

# サブプロットを設定し、各描画メソッドでグラフを描画
plt.figure(figsize=(15, 10))  # 図全体のサイズを指定

# Circular Layoutでグラフを描画
plt.subplot(231)  # 2行3列のグリッドの1番目の位置にプロットを作成
nx.draw_circular(G, node_size=40, node_color='red', with_labels=False)  # Circular Layoutでネットワークを描画
plt.title('Circular Layout')  # グラフのタイトルを設定

# Kamada-Kawai Layoutでグラフを描画
plt.subplot(232)  # 2行3列のグリッドの2番目の位置にプロットを作成
nx.draw_kamada_kawai(G, node_size=40, node_color='red', with_labels=False)  # Kamada-Kawai Layoutでネットワークを描画
plt.title('Kamada-Kawai Layout')  # グラフのタイトルを設定

# Random Layoutでグラフを描画
plt.subplot(233)  # 2行3列のグリッドの3番目の位置にプロットを作成
nx.draw_random(G, node_size=40, node_color='red', with_labels=False)  # Random Layoutでネットワークを描画
plt.title('Random Layout')  # グラフのタイトルを設定

# Spectral Layoutでグラフを描画
plt.subplot(234)  # 2行3列のグリッドの4番目の位置にプロットを作成
nx.draw_spectral(G, node_size=40, node_color='red', with_labels=False)  # Spectral Layoutでネットワークを描画
plt.title('Spectral Layout')  # グラフのタイトルを設定

# Spring Layoutでグラフを描画
plt.subplot(235)  # 2行3列のグリッドの5番目の位置にプロットを作成
nx.draw_spring(G, node_size=40, node_color='red', with_labels=False)  # Spring Layoutでネットワークを描画
plt.title('Spring Layout')  # グラフのタイトルを設定

# Shell Layoutでグラフを描画
plt.subplot(236)  # 2行3列のグリッドの6番目の位置にプロットを作成
nx.draw_shell(G, node_size=40, node_color='red', with_labels=False)  # Shell Layoutでネットワークを描画
plt.title('Shell Layout')  # グラフのタイトルを設定

# 描画した図を表示
plt.tight_layout()  # サブプロット間のレイアウトを自動調整
plt.show()  # 図を表示
```

## 特徴量の計算
### 頂点や辺の数など、ネットワークの構造的特徴を表す特微量
ネットワークの構造的特徴量の計算は、ネットワークの性質や挙動を理解し、解析するための重要なステップです。この計算にはいくつかの重要な意義があります。<br>

1. **ネットワークの全体的な構造を理解する**<br>
特徴量の計算を通じてネットワークの全体的な構造を把握できます。例えば、**ノード数**や**エッジ数**、**密度**などの基本的な指標は、ネットワークの規模や接続の度合いを示し、ネットワークがどのように構築されているかを理解するのに役立ちます。<br>

2. **ノードの重要度や影響力を評価する**<br>
**中心性指標**（次数中心性、固有ベクトル中心性、近接中心性、媒介中心性）を計算することで、ネットワーク内の各ノードの重要度や影響力を評価できます。これにより、ネットワーク内で重要な役割を果たすノードを特定し、リソースの割り当てやターゲット戦略の設計に役立てることができます。<br>

3. **ネットワークのローカルなパターンや構造を分析する**<br>
**クラスタリング係数**や**連結成分**の計算を通じて、ネットワーク内のローカルなパターンや構造を分析できます。これにより、ノードがどの程度互いに連結しているか、ネットワーク内にどのようなサブグループが存在するかを理解することができます。<br>

4. **ネットワークの効率性や耐障害性を評価する**<br>
**平均経路長**や**直径**の計算により、ネットワークの効率性や耐障害性を評価できます。短い経路長は効率的な情報伝達を示し、長い直径はネットワークの広がりを示します。これにより、ネットワークがどれだけ迅速かつ効率的に情報を伝達できるか、またどの程度の障害に耐えられるかを評価することができます。<br>

5. **比較とベンチマーク**<br>
異なるネットワーク同士を比較する際に特徴量の計算が役立ちます。例えば、異なるソーシャルネットワークの構造や特性を比較することで、それぞれのネットワークの特徴や強み、弱みを明らかにすることができます。また、ネットワークの進化や変化を追跡する際にも、時間ごとの特徴量の変化を観察することでネットワークのダイナミクスを理解することができます。<br>

6. **モデルの検証と改善**<br>
ネットワークモデルの検証や改善にも特徴量の計算が重要です。実際のネットワークと理論的なモデルの特徴量を比較することでモデルの適合度を評価し、モデルの改良点を見つけることができます。<br>


このスクリプトは、NetworkXライブラリを使用して様々なグラフを生成し、それらを可視化するものです。最初に、追加のラベルを表示する関数add_features_to_plotが定義されており、この関数はグラフの各ノードの特徴量（この場合、ノードの次数）をラベルとしてノードの隣に表示します。プロットのサイズを指定し、サブプロットとして各グラフを描画します。

10個の頂点と確率0.3でエッジを持つランダムグラフを生成し、円形レイアウトで描画します。また、Petersenグラフ、10個の頂点からなるサイクルグラフ、10個の頂点からなる完全グラフ、5対5の完全二部グラフ、バーベルグラフ、1つの中心ノードと9つの周辺ノードからなるスターグラフ、10個の頂点からなるホイールグラフ、およびバラバシ・アルバートモデルに基づくグラフをそれぞれ異なるレイアウトで描画します。

```python
import networkx as nx
import matplotlib.pyplot as plt

# 追加のラベルを表示する関数
def add_features_to_plot(G, pos, ax, features, vertical=True):
    labels = {node: f"{features[node]}" for node in G.nodes()}
    for node, (x, y) in pos.items():
        if vertical:
            ax.text(x, y + 0.1, labels[node], fontsize=12, ha='center', va='bottom', color='blue', fontweight='bold')
        else:
            ax.text(x + 0.1, y, labels[node], fontsize=12, ha='left', va='center', color='blue', fontweight='bold')

# プロットのサイズを指定
plt.figure(figsize=(15, 15))

# 10個の頂点と確率0.3でエッジを持つランダムグラフを生成し、円形レイアウトで描画します。
rnd = nx.gnp_random_graph(10, 0.3)
pos = nx.circular_layout(rnd)
plt.subplot(331)
nx.draw_networkx_nodes(rnd, pos, node_size=200, node_color='red')
nx.draw_networkx_edges(rnd, pos)
nx.draw_networkx_labels(rnd, pos, font_weight='bold')
plt.title('Random Graph (10 nodes, p=0.3)')
add_features_to_plot(rnd, pos, plt.gca(), nx.degree(rnd), vertical=True)

# Petersenグラフをシェルレイアウトで描画します。
petersen = nx.petersen_graph()
pos = nx.shell_layout(petersen, nlist=[range(5, 10), range(5)])
plt.subplot(332)
nx.draw_networkx_nodes(petersen, pos, node_size=200, node_color='red')
nx.draw_networkx_edges(petersen, pos)
nx.draw_networkx_labels(petersen, pos, font_weight='bold')
plt.title('Petersen Graph')
add_features_to_plot(petersen, pos, plt.gca(), nx.degree(petersen), vertical=True)

# 10個の頂点からなるサイクルグラフを円形レイアウトで描画します。
cycle = nx.cycle_graph(10)
pos = nx.circular_layout(cycle)
plt.subplot(333)
nx.draw_networkx_nodes(cycle, pos, node_size=200, node_color='red')
nx.draw_networkx_edges(cycle, pos)
nx.draw_networkx_labels(cycle, pos, font_weight='bold')
plt.title('Cycle Graph (10 nodes)')
add_features_to_plot(cycle, pos, plt.gca(), nx.degree(cycle), vertical=True)

# 10個の頂点からなる完全グラフを円形レイアウトで描画します。
K_10 = nx.complete_graph(10)
pos = nx.circular_layout(K_10)
plt.subplot(334)
nx.draw_networkx_nodes(K_10, pos, node_size=200, node_color='red')
nx.draw_networkx_edges(K_10, pos)
nx.draw_networkx_labels(K_10, pos, font_weight='bold')
plt.title('Complete Graph (10 nodes)')
add_features_to_plot(K_10, pos, plt.gca(), nx.degree(K_10), vertical=True)

# 5対5の完全二部グラフを円形レイアウトで描画します。
K_5_5 = nx.complete_bipartite_graph(5, 5)
pos = nx.circular_layout(K_5_5)
plt.subplot(335)
nx.draw_networkx_nodes(K_5_5, pos, node_size=200, node_color='red')
nx.draw_networkx_edges(K_5_5, pos)
nx.draw_networkx_labels(K_5_5, pos, font_weight='bold')
plt.title('Complete Bipartite Graph (5, 5)')
add_features_to_plot(K_5_5, pos, plt.gca(), nx.degree(K_5_5), vertical=True)

# バーベルグラフをバネモデルで描画します。
barbell = nx.barbell_graph(5, 5)
pos = nx.spring_layout(barbell)
plt.subplot(336)
nx.draw_networkx_nodes(barbell, pos, node_size=200, node_color='red')
nx.draw_networkx_edges(barbell, pos)
nx.draw_networkx_labels(barbell, pos, font_weight='bold')
plt.title('Barbell Graph (m1=5, m2=5)')
add_features_to_plot(barbell, pos, plt.gca(), nx.degree(barbell), vertical=False)

# 1つの中心ノードと9つの周辺ノードからなるスターグラフを描画します。
star = nx.star_graph(9)
pos = nx.spring_layout(star)
plt.subplot(337)
nx.draw_networkx_nodes(star, pos, node_size=200, node_color='red')
nx.draw_networkx_edges(star, pos)
nx.draw_networkx_labels(star, pos, font_weight='bold')
plt.title('Star Graph (9 peripheral nodes)')
add_features_to_plot(star, pos, plt.gca(), nx.degree(star), vertical=False)

# 10個の頂点からなるホイールグラフを描画します。
wheel = nx.wheel_graph(10)
pos = nx.spring_layout(wheel)
plt.subplot(338)
nx.draw_networkx_nodes(wheel, pos, node_size=200, node_color='red')
nx.draw_networkx_edges(wheel, pos)
nx.draw_networkx_labels(wheel, pos, font_weight='bold')
plt.title('Wheel Graph (10 nodes)')
add_features_to_plot(wheel, pos, plt.gca(), nx.degree(wheel), vertical=False)

# バラバシ・アルバートモデルに基づくグラフをバネモデルで描画します。
ba = nx.barabasi_albert_graph(10, 2)
pos = nx.spring_layout(ba)
plt.subplot(339)
nx.draw_networkx_nodes(ba, pos, node_size=200, node_color='red')
nx.draw_networkx_edges(ba, pos)
nx.draw_networkx_labels(ba, pos, font_weight='bold')
plt.title('Barabási-Albert Graph (n=10, m=2)')
add_features_to_plot(ba, pos, plt.gca(), nx.degree(ba), vertical=False)

plt.tight_layout(pad=3.0)
plt.show()
```

## ネットワークの密度とクラスタ係数の計算
ネットワークの密度とクラスタ係数は、ネットワークの構造を表す重要な特徴量です。

ネットワークの密度は、実際に接続されたエッジの数と、可能な最大のエッジの数の比率を表します。密度が高いほどノード同士が密に繋がっており、情報や影響が素早く広がりやすくなります。ネットワークのノード数 |V| とエッジ数 |E| を用いて、密度 D は以下の式で計算されます。

$D = \frac{2|E|}{|V|(|V| - 1)}$

一方、クラスタ係数は、ネットワークの構造を表す特徴量の一つです。クラスタ係数は、ある頂点とその隣接頂点の間に存在する実際のエッジの数と、理論上の最大エッジ数の比率を表す指標です。クラスタ係数の値は0から1の範囲をとり、値が高いほど頂点とその周辺が密接に結びついていることを示します。

クラスタ係数の値から、ネットワークの性質を推測することが可能です。例えば、クラスタ係数が0に近い場合はネットワークがランダムに構成されている傾向があり、1に近い場合はネットワークが高度にクラスタリングされていることを示します。実際のネットワークにおいては、社会ネットワークではクラスタ係数が高く、電力グリッドネットワークでは低い傾向にあります。

このスクリプトでは、NetworkXライブラリのクラスタ係数計算関数を使用し、各グラフのクラスタ係数の平均値をプロット上に表示します。これにより、ランダムグラフやネットワークモデルなどの異なるグラフ構造がどの程度クラスタリングされているかを比較することができます。

ノード i のクラスタ係数 Ci は、そのノードの近傍ノード間の実際のエッジ数 ei とそのノードの次数（接続数） ki を用いて以下の式で計算されます。

$C_i = \frac{2 e_i}{k_i (k_i - 1)}$

これらの指標は、ネットワークの構造や特性を定量化し、異なるタイプのネットワークを比較や分析するために広く利用されます。

## ネットワークの密度とクラスタリング係数を可視化
このスクリプトはネットワークグラフを生成し、各グラフの密度と各グラフ全体のクラスタリング係数を計算して可視化するためのものです。PythonのNetworkXライブラリとMatplotlibを使用して複数の種類のグラフを描画します。各グラフは異なるレイアウトで表示され、ノード、エッジ、ラベルが描画されます。タイトルにはネットワークの密度とクラスタリング係数が表示されます。

```python
import networkx as nx
import matplotlib.pyplot as plt

# ネットワークを描画する関数を定義
def draw_network_with_metrics(ax, G, pos):
    # ノードを描画
    nx.draw_networkx_nodes(G, pos, node_size=200, node_color='red', ax=ax)
    # ラベルを描画
    nx.draw_networkx_labels(G, pos, font_weight='bold', ax=ax)
    # エッジを描画
    nx.draw_networkx_edges(G, pos, ax=ax)
    # グラフの密度とクラスタリング係数を計算
    density = nx.density(G)
    clustering_coefficient = nx.average_clustering(G)
    # タイトルとして表示
    ax.set_title('Density: {:.5f}\nClustering Coefficient: {:.5f}'.format(density, clustering_coefficient))
    # 軸をオフにする
    ax.axis('off')

# サブプロットを作成
fig, axs = plt.subplots(3, 3, figsize=(15, 15))
axs = axs.flatten()

# 各グラフを生成し、描画関数を呼び出す
rnd = nx.gnp_random_graph(10, 0.3)
pos_rnd = nx.circular_layout(rnd)
draw_network_with_metrics(axs[0], rnd, pos_rnd)

petersen = nx.petersen_graph()
pos_petersen = nx.shell_layout(petersen, nlist=[range(5, 10), range(5)])
draw_network_with_metrics(axs[1], petersen, pos_petersen)

cycle = nx.cycle_graph(10)
pos_cycle = nx.circular_layout(cycle)
draw_network_with_metrics(axs[2], cycle, pos_cycle)

K_10 = nx.complete_graph(10)
pos_K_10 = nx.circular_layout(K_10)
draw_network_with_metrics(axs[3], K_10, pos_K_10)

K_5_5 = nx.complete_bipartite_graph(5, 5)
pos_K_5_5 = nx.shell_layout(K_5_5, nlist=[range(5, 10), range(5)])
draw_network_with_metrics(axs[4], K_5_5, pos_K_5_5)

barbell = nx.barbell_graph(5, 5)
pos_barbell = nx.spring_layout(barbell)
draw_network_with_metrics(axs[5], barbell, pos_barbell)

star = nx.star_graph(9)
pos_star = nx.spring_layout(star)
draw_network_with_metrics(axs[6], star, pos_star)

wheel = nx.wheel_graph(10)
pos_wheel = nx.spring_layout(wheel)
draw_network_with_metrics(axs[7], wheel, pos_wheel)

ba = nx.barabasi_albert_graph(10, 2)
pos_ba = nx.spring_layout(ba)
draw_network_with_metrics(axs[8], ba, pos_ba)

# 描画
plt.tight_layout()
plt.show()
```

## 中心性の計算
中心性の計算は、ネットワーク内で特に**重要な役割を果たすノードを見つける**ための手法です。ネットワークにおいて特に重要な役割とは、例えば情報の迅速な伝達や異なる部分の結びつきを促進するなど、ネットワーク全体の機能や構造に大きな影響を与えるポジションや役割を指します。中心性の指標を用いることで重要なノードを定量的に評価し、理解することが可能です。<br>
<br>
いくつかの代表的な中心性指標とその意味を説明します。<br>
1. **次数中心性 (Degree Centrality)**<br>
・各ノードが**直接つながっている他のノードの数**を示します。次数が高いノードは、他の多くのノードと直接つながっているため、情報の伝達やネットワークの構造に大きな影響を与える可能性があります。<br>
2. **固有ベクトル中心性 (Eigenvector Centrality)**<br>
・ノードの重要性を**ノードに接続している他のノードの重要性**も考慮して評価します。単に接続数が多いだけでなく、重要なノードとつながっているノードが高い中心性を持つことを意味します。<br>
3. **近接中心性 (Closeness Centrality)**<br>
・あるノードから他のすべてのノードまでの**平均最短経路長**を測ります。近接中心性が高いノードは、ネットワーク内の他のすべてのノードに迅速に到達できるため、情報の流通において効率的な位置にあると言えます。<br>
4. **媒介中心性 (Betweenness Centrality)**<br>
・ノードがネットワーク内の他のノード間の最短経路上にどれだけ多く現れるかを測定します。媒介中心性が高いノードは異なる部分のネットワークをつなぐ**橋渡し**（ハブ）の役割を果たし、ネットワークの統合性や情報の流れに重要な影響を与えます。<br>
<br>
### 中心性の計算の意義
このスクリプトは空手クラブのデータセットから各ノードの次数を求め、いくつかの中心性の定義から求められた数値を基にネットワークの繋がり（中心性）の強さを階層的に視覚化したものです。中心性の定義の違いによるグラフの変化を直感的に理解できます。<br>
中心性の計算は、以下のような様々な実際の応用に役立ちます。<br>

**ソーシャルネットワーク分析：** 重要な人物（インフルエンサー）を特定する。<br>
**インターネットのトラフィック管理：** 主要なハブやボトルネックを特定し、効率的なデータ転送を実現する。<br>
**病原体の拡散予測：** 疾病の伝播を防ぐためのキーポイントを特定する。<br>
**組織の管理：** 組織内の重要な連絡点や情報の流通を把握する。<br>

中心性の指標を使ってネットワークを解析することで、重要なノードを特定し、ネットワーク全体の構造と機能を深く理解することができます。<br>
※nx.draw_networkx_nodes 関数は norm 引数を受け取りません。そのため、カラースケールの正規化はカラーマップの設定で行っています。

```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import os

# 描画関数を定義
def draw_h(G, pos, measures, measure_name, ax):
    """
    指定された中心性指標に基づいてネットワークグラフを描画します。

    Parameters:
    G (networkx.Graph): 描画対象のネットワークグラフ。
    pos (dict): ノードの配置を示す辞書。
    measures (dict): ノードの中心性指標を格納する辞書。
    measure_name (str): 描画する中心性指標の名前。
    ax (matplotlib.axes.Axes): グラフを描画するサブプロットのAxesオブジェクト。

    描画手順:
    1. ノードの描画: 指定された中心性指標の値に基づいてノードの色を設定し、ノードを描画します。
    2. ノードの色設定: 対数スケールで色を正規化します。
    3. ノードのラベル描画: ノードにラベルを描画します。
    4. エッジの描画: ノード間のエッジを描画します。
    5. タイトルと軸の設定: グラフのタイトルを設定し、軸を非表示にします。
    """
     # 正規化
    vmin = min(measures.values())
    vmax = max(measures.values())
    
    nodes = nx.draw_networkx_nodes(G, pos, node_size=250,
                                  cmap=plt.cm.plasma,
                                  node_color=list(measures.values()),
                                  nodelist=list(measures.keys()),
                                  ax=ax)
    
    edges = nx.draw_networkx_edges(G, pos, ax=ax)
    ax.set_title(measure_name)
    ax.axis('off')
    
    # カラーバーの追加
    sm = plt.cm.ScalarMappable(cmap=plt.cm.plasma, norm=plt.Normalize(vmin=vmin, vmax=vmax))
    sm.set_array([])
    cbar = plt.colorbar(sm, ax=ax)
    
    nodes = nx.draw_networkx_nodes(G, pos, node_size=250,
                                  cmap=plt.cm.plasma,
                                  node_color=list(measures.values()),
                                  nodelist=list(measures.keys()), ax=ax)
    nodes.set_norm(mcolors.SymLogNorm(linthresh=0.01, linscale=1))
    nx.draw_networkx_labels(G, pos, font_size=10, font_color='black', font_family='sans-serif', ax=ax)
    edges = nx.draw_networkx_edges(G, pos, ax=ax)
    ax.set_title(measure_name)
    ax.axis('off')

# カレントディレクトリを変更
new_directory = r"D:\Python\NetworkX\lesmis"
os.chdir(new_directory)

# ファイルパスを指定
file_path = r'lesmis.gml'

# ファイルを読み込んでグラフを描画
G = nx.read_gml(file_path)
pos = nx.spring_layout(G)

# 描画エリアのサイズを設定して各中心性指標を計算し、描画
plt.figure(figsize=(18, 18))
ax1 = plt.subplot(2, 2, 1)
draw_h(G, pos, nx.degree_centrality(G), 'Degree Centrality', ax1)

ax2 = plt.subplot(2, 2, 2)
draw_h(G, pos, nx.eigenvector_centrality(G), 'Eigenvector Centrality', ax2)

ax3 = plt.subplot(2, 2, 3)
draw_h(G, pos, nx.closeness_centrality(G), 'Closeness Centrality', ax3)

ax4 = plt.subplot(2, 2, 4)
draw_h(G, pos, nx.betweenness_centrality(G), 'Betweenness Centrality', ax4)

plt.tight_layout()
plt.show()
```

# コミュニティの抽出
多くのネットワークには、社会ネットワークにおける友人グループのように、密に結びついた頂点の集まりが存在します。このような密な部分ネットワークをコミュニティと呼びます。例えば、SNSにおけるコミュニティは、似た興味や関心を持つユーザーグループに対応します。

コミュニティ抽出では、類似した頂点が集まり、それがひとつのコミュニティを形成することが多いです。これにより、SNSでは興味関心が似ているユーザーに関連する内容の投稿をタイムラインに表示する際に、有用な情報をシステムに提供することができます。また、グループ内で中心的な頂点や他のグループとの境界にある頂点を特定することは、情報伝播における影響力や伝播の範囲を理解する上で重要です。

たとえば『レ・ミゼラブル』の登場人物たちのネットワークを考えてみましょう。物語の中で密に関わり合っているキャラクターたちのグループを見つけ出すことで、革命家たちの集団や家族の関係など、物語の中の隠れた構造や重要なグループを発見することができます。これにより、物語の理解が深まり、キャラクター同士の関係や物語の進行における重要なポイントを把握する手助けとなります。

コミュニティ抽出の手法は多岐にわたり、抽出するコミュニティの大きさや個数は事前に定まっていない場合があります。また、コミュニティの定義にも複数のアプローチがあります。

### 類似したエッジが構成するコミュニティを抽出する

このスクリプトはネットワーク内のコミュニティを抽出し、可視化する手順を示しています。<br>
以下の手順でネットワークのコミュニティ抽出と可視化を行っています。<br>
1. **ライブラリのインポート**<br>
スクリプトの最初で必要なライブラリをインポートしています。これにはnetworkx（ネットワーク分析のためのライブラリ）、matplotlib.pyplot（グラフ描画のためのライブラリ）、community（コミュニティ抽出アルゴリズム）、pandas（データ処理のためのライブラリ）が含まれます。<br>

2. **GMLファイルの読み込み**<br>
lesmis.gmlというファイルパスから、GML形式のネットワークデータを読み込んでいます。GML形式は、ネットワーク構造を記述するための一般的な形式です。<br>

3. **ノード属性のDataFrame作成**<br>
読み込んだネットワークから、ノードの属性を持つDataFrameを作成しています。ここでは、各ノードのIDと初期値として0を持つmyvalueという属性を設定しています。<br>

4. **コミュニティの抽出**<br>
**Girvan-Newmanアルゴリズム**を使用して、ネットワークからコミュニティ（グループ）を抽出しています。このアルゴリズムは、ネットワーク内の密な部分を特定し、それらをコミュニティとして分割します。<br>

5. **ネットワークの可視化**<br>
抽出したコミュニティ番号を各ノードの属性に設定し、nx.kamada_kawai_layoutを使用してノードのレイアウト（配置）を設定しています。このレイアウトは、ノード間の距離を少し広げるために使用されています。最後にmatplotlibを使用して、ノードの色をコミュニティ番号に基づいて描画し、ラベル付きでネットワークを表示しています。<br>
<br>
このスクリプトを実行することで、ネットワーク内のコミュニティ構造を抽出し、可視化することができます。<br>
※コミュニティ抽出の調整や、カラーマップの調整を行っていないため適切な可視化ではありません！

```python
# 必要なライブラリをインポート
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import community
import pandas as pd

# GMLファイルのパス
gml_file = r'D:\Python\NetworkX\lesmis\lesmis.gml'

# GML形式のファイルを読み込み
G = nx.read_gml(gml_file)

# ノード属性を持つDataFrameを作成
carac = pd.DataFrame({'ID': G.nodes(), 'myvalue': [0] * len(G.nodes())})

# Girvan-Newmanアルゴリズムでコミュニティを抽出
communities_generator = community.girvan_newman(G)
# コミュニティを取得
top_level_communities = next(communities_generator)
# 次のレベルのコミュニティを取得
next_level_communities = next(communities_generator)

# コミュニティ番号をノード属性に設定してネットワークを描画
for m in range(len(next_level_communities)):
    for n in next_level_communities[m]:
        carac.loc[carac.ID == n, 'myvalue'] = m

# レイアウトの設定（ノード間の距離を少し広げる）
pos = nx.kamada_kawai_layout(G)

# ネットワークを描画
plt.figure(figsize=(10, 8))
nx.draw(G, pos, node_color=carac['myvalue'], node_size=120, with_labels=True, cmap=plt.cm.gist_ncar)
plt.show()
```

例えば、最初の分析では全体的なネットワークの構造や特徴を把握することができ、次にそのネットワークの密度やクラスタ係数を算出して、特定の部分構造や局所的な性質を理解します。さらには、重要なノードの特定や、ネットワーク内の影響関係の解明など、分析の目的に応じて深掘りしていくことが重要です。
